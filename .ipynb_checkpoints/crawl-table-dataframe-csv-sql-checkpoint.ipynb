```python
# Import packages
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
import csv
```

```python
# Set url, request, make request
url = "http://www.nfl.com/stats/categorystats?archive=false&conference=null&statisticCategory=PASSING&season=2014&seasonType=REG&experience=&tabSeq=0&qualified=false&Submit=Go"
response = requests.get(url)
html = response.content
```

```python
# Set bs object
soup = bs(html, 'lxml')
table = soup.find('table', attrs={'id': 'result'})
```

```python
# Clean data
rows = []

for row in table.findAll('tr'):
    cells = []
    for cell in row.findAll('td'):
        text = cell.text.replace('&nbsp;', '').strip()
        cells.append(text)
    rows.append(cells)
```

```python
# Define columns create DataFrame object
columns = ["Rk", "Player", "Team", "Pos", "Comp", "Att", "Pct", "Att/G", "Yds", "Avg", "Yds/G", "TD", "Int", "1st", "1st%", "Lng", "20+", "40+", "Sck", "Rate"]
df = pd.DataFrame(rows, columns=columns)
```
